###
## Column Droping:
###

```python

import pandas as pd

a = pd.read_csv('colors(csvfile5).csv')
a

b = a.drop(['Cars'],axis = 1) # or use ===> a.drop(['Cars'],axis =1, inplace = True) as it will modify the orignal data
b

c = a.drop(['Cars','float'],axis = 1)
c

a.drop(['Cars'],axis =1, inplace = True)
a

```
