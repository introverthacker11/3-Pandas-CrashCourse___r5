```python
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv')
a
```

```python
# this'll give mean,max,min,count,sum of column rank a/c to each college
a.groupby(['College','Department']).agg({'Rank' : ['mean','max','min','count','sum']}) 
```

```python
# this'll give mean,max,min,count,sum of column rank a/c to each department
a.groupby(['Department','College']).agg({'Rank' : ['mean','max','min','count','sum']}) 
```

```python
# this'll give mean,max,min,count,sum of column rank,height,age a/c to each College
a.groupby(['College','Department']).agg({'Rank' : ['mean','max','min','count','sum'],'Height(foot.inch)' : ['mean','max','min','count','sum'],'Age' : ['mean','max','min','count','sum']}) 
```

```python
# this'll give mean,max,min,count,sum of column rank,height,age a/c to each department
a.groupby(['Department','College']).agg({'Rank' : ['mean','max','min','count','sum'],'Height(foot.inch)' : ['mean','max','min','count','sum'],'Age' : ['mean','max','min','count','sum']})
```
