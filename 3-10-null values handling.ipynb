```python
import pandas as pd

a = pd.read_csv('null(csvfile6).csv')
a
a.isnull()
a.isnull().sum()
a.notnull()
a.notnull().sum()
a.fillna(0) # it will temporarily fill null values to 0
a.fillna('Not Known') # same as line 9
a.fillna('Not Known', inplace = True) # it will fill null values to not known for this cell only
a.info()
a
```
